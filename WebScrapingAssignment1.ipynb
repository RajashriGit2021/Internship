# Q1.Write a python program to display all the header tags from‘en.wikipedia.org/wiki/Main_Page’.

# Q2.Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
from bs4 import BeautifulSoup
import requests
import re
 
# Downloading imdb top 100 movie's data
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
movies = soup.select('td.titleColumn')
#crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value')
           for b in soup.select('td.posterColumn span[name=ir]')]
list = []
i=1
# create a empty list for storing movie information
list = []
 
for index in range(0, 100):
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    data = {"movie_title": movie_title,
            "year": year,"rating": ratings[index] }
    list.append(data)
 
# printing movie details with its rating.
print('No:       Movie Title          Rating       Release Year')
for movie in list:
    print(i,': ',movie['movie_title'],'-',movie['rating'], '-('+movie['year'] +
          ')')
    i=i+1
```

# Q4.Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
from bs4 import BeautifulSoup
import requests
import re
 
url = 'https://www.bookpage.com'
page = requests.get(url)
page.text

soup=BeautifulSoup(page.content)
f1_title=soup.find_all('h4',class_="italic")


book_name=[]
for i in f1_title:
    for j in i.find_all("a"):
        book_name.append(j.text.replace("\n"," "))

AuthorName=soup.find_all('p',class_="sans bold")

Author_name=[]
for i in AuthorName:
    Author_name.append(i.text.replace("\n"," "))
Review=soup.find_all('p',class_="excerpt")

Reviews=[]
for i in Review:
    Reviews.append(i.text.replace("\n"," "))

genere=soup.find_all('p',class_="genre-links hidden-phone")
generes=[]
for i in genere:
    generes.append(i.text.replace("\n"," "))

import pandas as pd
book_series=pd.DataFrame({})
book_series['book_name']=book_name
book_series['Author_name']=Author_name
book_series['Reviews']=Reviews
book_series = book_series.iloc[1: , :]
book_series['generes']=generes
left_book_series = book_series.style.set_properties(**{'text-align': 'left'})
display(left_book_series)
```

# Q7 Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
}
url = 'https://www.amazon.in/s?k=mobile+phone+under+20000+rupees&ref=nb_sb_noss_1'
page = requests.get(url, headers=headers)
soup=BeautifulSoup(page.content)
page
```

```python
title1=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")
title=[]
for i in title1:
        title.append(i.text.replace("\n"," "))
price1=soup.find_all('span',class_="a-price-whole")
price=[]
for i in price1:
    price.append(i.text.replace("\n"," "))
#price.append("NA")

img1=[]
images = soup.find_all('img', {'src':re.compile('.jpg')})
for image in images: 
    img1.append(image['src'])
img1=img1[:16]

rating1=soup.find_all('span',class_="a-icon-alt")
rating=[]
for i in rating1:
    rating.append(i.text.replace("\n"," "))
rating=rating[:16]
Amazondata=pd.DataFrame({})
Amazondata['title']=title
Amazondata['price']=price
Amazondata['rating']=rating
Amazondata['url']=img1
left_Amazondata = Amazondata.style.set_properties(**{'text-align': 'left'})
display(left_Amazondata)
```

# Q8.Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

url = 'https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YSEtdI4zaHs'
page = requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)

period1=soup.find_all('p',class_="period-name")
period=[]
for i in period1:
        period.append(i.text.replace("\n"," "))

Short_description1=soup.find_all('p',class_="short-desc")
Short_description=[]
for i in Short_description1:
    Short_description.append(i.text.replace("\n"," "))
    
description1=soup.find_all('div',class_="col-sm-10 forecast-text")
description=[]
for i in description1:
    description.append(i.text.replace("\n"," "))
description=description[:9]    
temperature11=soup.find_all('p',class_="temp temp-low")
temperature1=[]
temperature2=[]
for i in temperature11:
    temperature1.append(i.text.replace("\n"," "))
temperature11=soup.find_all('p',class_="temp temp-high")
for i in temperature11:
    temperature2.append(i.text.replace("\n"," "))

temp = [None]*(len(temperature1)+len(temperature2))
temp[::2] = temperature2
temp[1::2] = temperature1


weather=pd.DataFrame({})
weather['period']=period
weather['Short_description']=Short_description1
weather['temperature']=temp
weather['description']=description
left_weather = weather.style.set_properties(**{'text-align': 'left'})
display(left_weather)
```

# 一	9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
page = requests.get(url)
soup=BeautifulSoup(page.content)

title1=soup.find_all('div',class_="heading_4_5 profile")
title=[]
for i in title1:
        title.append(i.text.replace("\n"," "))

company1=soup.find_all('div',class_="heading_6 company_name")
company=[]
for i in company1:
    company.append(i.text.replace("\n"," "))

CTC1=soup.find_all('div',class_="item_body")
CTC=[]
for i in CTC1:
    CTC.append(i.text.replace("\n"," "))
CTC.pop(0)
CTC=CTC[::3]

Adate1=soup.find_all('div',class_="item_body")
Adate=[]
for i in Adate1:
    Adate.append(i.text.replace("\n"," "))
Adate.pop(0)
Adate.pop(0)
Adate=Adate[::3] 

import pandas as pd
JobDetails=pd.DataFrame({})
JobDetails['title']=title
JobDetails['company']=company
JobDetails['CTC']=CTC
JobDetails['Adate']=Adate
left_JobDetails = JobDetails.style.set_properties(**{'text-align': 'left'})
display(left_JobDetails)
```
